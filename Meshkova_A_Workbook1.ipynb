Задание 4.3.1

```python
from statistics import *
from random import random
import matplotlib.pyplot as plt
from numpy import arange

line = arange(0, 1, 0.1)
numbers = [random() for i in range(10)]
avrg = mean(numbers)
med = median(numbers)
print('Среднее значение —', avrg, '\nМедиана —', med)
plt.show()
plt.grid()
plt.scatter(line, numbers)
```

```python
from numpy import *
import matplotlib.pyplot as plt

x = arange(1, 11, 1)
f = sqrt(1 + exp(sqrt(x)) + cos(x ** 2)) / abs(1 - (sin(x) ** 3)) + log(abs(2 * x))
plt.grid()
plt.plot(x, f)
y, d = x[:5], f[:5]
plt.show()
plt.grid()
plt.scatter(y, d)
```

```python
import matplotlib.pyplot as plt
from numpy import *

x = arange(1, 10, 1)
func = abs(cos(x * exp(cos(x) + log(x + 1))))
plt.grid()
plt.plot(x, func)
plt.fill_between(x, func)
field = trapz(func)
print(field)
```
